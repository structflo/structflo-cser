## PDF example with HungarianMatcher

```python
from structflo.cser.pipeline import ChemPipeline, HungarianMatcher

pipeline = ChemPipeline(
    weights="../runs/labels_detect/yolo11l_panels/weights/best.pt",
    tile=False,
    conf=0.20,
    matcher=HungarianMatcher(max_distance=500),
    )
```

```python
all_pages = pipeline.process_pdf("notebook-data/example.pdf")

for page_num, pairs in enumerate(all_pages):
    print(f"Page {page_num + 1}: {len(pairs)} compound pairs")
    for pair in pairs:
        print(f"  {pair.label_text:20s}  {pair.smiles}")
```

```python
pipeline.process_pdf("notebook-data/example.pdf", output_pdf="notebook-data/example-annotated.pdf")
```
